```python
# Leitura do arquivo csv
import pandas as pd
import numpy as np

teses = pd.read_csv("../data/teses.csv")
teses.head()
```

```python
# Verificando a quantidade de linhas: 15649
n=len(teses)
```

```python
# Adicionando a coluna nid (meu range: 10001-30000)
nids = range(10001,10001+n)
teses['nid'] = pd.Series(nids)
teses.head()
```

```python
# Nomes das colunas
teses.columns
```

```python
# Renomeando colunas 
teses = teses.rename(columns = {
    'nid'          : 'nid', 
    'titulo'       : 'title',
    'ano'          : 'field_ano',
    'autor'        : 'field_autor',
    'departamento' : 'field_departamento',
    'grau'         : 'field_grau',
    'orientador'   : 'field_orientador',
    'paginas'      : 'field_paginas',
    'resumo'       : 'field_resumo',
    'tombo'        : 'field_tombo',
    'volumes'      : 'field_volumes',
    'area'         : 'field_area'
})
teses.head()
```

```python
# Removendo espaços desnecessários de algumas colunas
teses['field_grau'] = teses['field_grau'].str.strip()
teses['field_departamento'] = teses['field_departamento'].str.strip()
```

```python
# Quantidade de ocorrências de departamento
len(teses.field_departamento.unique()) # 332
```

```python
# qual ocorrência de departamento mais ocorre?
teses.field_departamento.value_counts().head(22)
```

```python
# essas 22 ocorrências representam quanto do universo 15649 ?
teses.field_departamento.value_counts().head(22).sum()
```
